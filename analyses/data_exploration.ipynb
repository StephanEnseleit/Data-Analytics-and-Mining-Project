# Data exploration / Figure generation

```python
import pandas as pd

data = pd.read_csv('../data/communities_processed.csv')
```

## Correlation Matrix

```python
import dataframe_image as dfi

corr = data.corr()
styled_corr = corr.style.background_gradient(cmap='coolwarm').set_precision(3)

dfi.export(styled_corr,"colored_correlation_matrix.png")
```

## Distribution Plot

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import math

fig = make_subplots(rows=4, cols=3, start_cell="top-left", subplot_titles = list(data.columns))
fig.update_layout(title='Distributions of selected socio-economic features and crime rate', title_x=0.5)

for index, feature in zip(range(0, 12), data):
    row_index = (index // 3) + 1
    col_index = (index % 3) + 1
    fig.add_trace(go.Histogram(x = data[feature],  name=feature), row =row_index, col = col_index)
    fig.update_annotations(font_size=12)
    fig.update_layout(showlegend=False)

fig.show()
```
